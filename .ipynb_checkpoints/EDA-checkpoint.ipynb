## EDA
Steps involved in eda-
1. Find misiing data
2. Relation between missing data and output feature
3. Numerical features
3.1 Temporal features and their relation with each other
3.2 other numerical features
4. Categorical feature

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
## Display all the columns of the dataframe

pd.pandas.set_option('display.max_columns',None)
```

```python
ds=pd.read_csv("train.csv")
```

```python
ds
```

```python
ds.head()
```

```python
feature_mv=[feature for feature in ds.columns if ds[feature].isnull().sum() >1]
for feature in feature_mv:
    print(f"{feature} have {np.round(ds[feature].isnull().mean(),4)*100}% ")
```

```python
for feature in feature_mv:
    data=ds.copy()
    data[feature]=np.where(data[feature].isnull(),1,0) #In feature having missinng value missing value will be replaced by 1 and not by 0
    
    data.groupby(feature)['SalePrice'].median().plot.bar(color=('r','g')) # Find relation between missing value and output
    plt.title(feature)
    plt.show()
```

```python
#Here we can see their is a relation between missing value and output so we will replace it with some thing usefull
```

## For numerical feature

```python
num_feature=[feature for feature in ds.columns if ds[feature].dtypes!='O']
print(f'Number of numerical feature in dataset {len(num_feature)}')
```

```python
ds[num_feature]
```

```python
ds['Id'].count() #Id is unique feature so it will have no effect on output feature
```

```python
#Temporal feature
```

```python
feature_year=[feature for feature in ds.columns if 'Yr' in feature or 'Year' in feature]
```

```python
for feature in feature_year:
    print(feature, ':',ds[feature].unique())
```

```python
ds.groupby('YrSold')['SalePrice'].median().plot() 
```

```python
# Since price is decresing with year lets take rest of temporal feature with difference to the year house sold
```

```python
for feature in feature_year:
    if feature!='YrSold':
        data=ds.copy()
        data[feature]=data['YrSold']-data[feature] 
        plt.scatter(data[feature],data['SalePrice'])
        plt.show()
```

 Lets visualise rest of numerical features
    
    1.Discrete
    
    2.Countnious 

```python
d_feature=[feature for feature in num_feature if len(ds[feature].unique())<25 and feature not in feature_year+['Id']]
print(f"Number of dicrete numerical feature in dataset {len(d_feature)}")
```

```python
for feature in d_feature:
    data=ds.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature)
    plt.show()
```
