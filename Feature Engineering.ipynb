```python
import numpy as np
import pandas as pd
```

```python
pd.pandas.set_option('display.max_columns', None)
```

```python
ds=pd.read_csv("train.csv")
```

```python
ds
```

```python
ds.head()
```

```python
m_feature=[feature for feature in ds.columns if ds[feature].isnull().sum()>1]
```

```python
# Lets change categorical feature with missing values
```

```python
catfeature_mv=[feature for feature in m_feature if ds[feature].dtype=='O']
for feature in catfeature_mv:
    print(f"{feature} have {np.round(ds[feature].isnull().mean(),4)*100}% ")
```

```python
#For categorical feature replace missing value with new label
for feature in catfeature_mv:
    ds[feature]=np.where(ds[feature].isnull(),'Missing',ds[feature])
```

```python
for feature in catfeature_mv:
    print(f"{feature} have {np.round(ds[feature].isnull().mean(),4)*100}% ")
```

```python
numfeature_mv=[feature for feature in m_feature if ds[feature].dtype!='O']
```

```python
for feature in numfeature_mv:
    print(f"{feature} have {np.round(ds[feature].isnull().mean(),4)*100}% ")
```

```python
# Replace numerical feature with missing value with median
```

```python
for feature in numfeature_mv:
    ds[feature+'nan']=np.where(ds[feature].isnull(),1,0) #Since their is relation between missing value and sale price lets add new coloumn
    ds[feature]=np.where(ds[feature].isnull(),ds[feature].median(),ds[feature])
```

```python
for feature in numfeature_mv:
    print(f"{feature} have {np.round(ds[feature].isnull().mean(),4)*100}% ")
```

```python
ds.head()
```

```python
# For temporal values let take all values wrt to yearsold
```

```python
feature_year=[]
```

```python
for feature in ['YearBuilt','YearRemodAdd','GarageYrBlt']:
        ds[feature]=ds['YrSold']-ds[feature] 
```

```python
ds.head(20)
```

```python
## Numerical feature 
```

```python
import numpy as np
num_features=['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']
for feature in num_features:
    ds[feature]=np.log(ds[feature])
```

```python
ds.head()
```

```python
cat_feature=[feature for feature in ds.columns if ds[feature].dtype=='O']
```

```python
cat_feature
```

```python
for feature in cat_feature:
    temp=ds.groupby(feature)['Id'].count()/len(ds)
    temp_df=temp[temp>0.01].index
    ds[feature]=np.where(ds[feature].isin(temp_df),ds[feature],'rar_v')
```

```python
ds.head(10)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for i in cat_feature:
    ds[i]=le.fit_transform(ds[i])
```

```python
ds.head(10)
```

# feature scaling 

```python
scal_feature=[feature for feature in ds.columns if feature not in ['Id','SalePrice']]
print("Number of scalable feature {}".format(len(scal_feature)))
```

```python
from sklearn.preprocessing import MinMaxScaler
scale=MinMaxScaler()
scale.fit_transform(ds[scal_feature])
```

```python
data = pd.concat([ds[['Id','SalePrice']].reset_index(drop=True),
                    pd.DataFrame(scale.fit_transform(ds[scal_feature]),columns=scal_feature)],
                    axis=1)
```

```python
data.head()
```

```python
data.to_csv("house.csv",index=False)
```
